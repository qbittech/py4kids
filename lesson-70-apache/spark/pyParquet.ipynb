```python
import pandas as pd
```

```python
df = pd.read_csv("crimes_1yr.csv")
```

```python
df.head()
```

```python
df['CASE#'].head()
```

```python
# write the dataframe out to a parquet file
df.to_parquet('crimes.snappy.parquet' ,engine='auto', compression='snappy')
```

```python
import pyarrow.parquet as pq
data = pq.read_pandas('crimes.snappy.parquet', columns=['CASE#', 'DATE OF OCCURRENCE', 'SECONDARY DESCRIPTION']).to_pandas()
```
